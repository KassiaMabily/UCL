```python
import math
import tabulate
import itertools
from datetime import datetime, timedelta

radius_earth = 6373.0

#===============================================================================
#================================ FUNÇÕES ======================================
#===============================================================================

def is_valid_coordenate(latitude, longitude):
  if (abs(latitude) >= 90):
    return False, f"\nLatidude ({latitude}) inválida"
  elif (abs(longitude) >= 180):
    return False, f"\nLongitude ({longitude}) inválida"
  else:
    return True, ""

def validated_input(text):
  texts = [ item.strip() for item in text.split(',') ]
  
  coordinates = []

  for text in texts:
    try:
      coordinates.append(float(text))
    except ValueError:
      print("\nInsira apenas números")
      return


  if (len(coordinates) == 2):
    # coordinates = [ float(value) for value in values ]
    is_valid, message = is_valid_coordenate(coordinates[0], coordinates[1])
    if (is_valid):
      return coordinates
    else:
      print(message)
      return
  else:
    print("\nInsira apenas 2 números separados por vírgula")
    return

    


def print_tabulate(data):
  # header = data[0].keys()
  header = ["Saída", "Destino", "Aluno", "Data e hora", "Distância (KM)"]
  rows = []
  for line in data:
    rows.append([ line["pin X"], line["pin Y"], line["name"], line["datetime"], line["distance"] ])
  
  print(tabulate.tabulate(rows, header))
  
def calculate_distance_haversine(pin_1, pin_2):
  latitude_1 = math.radians(pin_1[0])
  longitude_1 = math.radians(pin_1[1])

  latitude_2 = math.radians(pin_2[0])
  longitude_2 = math.radians(pin_2[1])

  distance_latitude = latitude_2 - latitude_1
  distance_longitude = longitude_2 - longitude_1

  formule_haversine = math.sin(distance_latitude / 2)**2 + math.cos(latitude_1) * math.cos(latitude_2) * math.sin(distance_longitude / 2)**2
  distance_final = 2 * math.atan2(math.sqrt(formule_haversine), math.sqrt(1 - formule_haversine))

  distance_final_km = radius_earth * distance_final;

  return round(distance_final_km, 2)

def selected_student(selected_address, addresses):
  excluded_adresses = []
  for address in addresses:
    if (address != selected_address):
      excluded_adresses.append(address)
      print(f"Aluno {address['name']} não adicionado na rota devido a um conflito de horário.")

  return excluded_adresses
  
def check_exists(y, routes):
  for route in routes:
    if ((route["pin X"] == [y[0], y[1]]) or (route["pin Y"] == [y[0], y[1]])):
      return True

  return False

def ordenate_data(addresses):
  final_route = []
  excluded_adresses = []
  # Endereço de partida
  next_x = addresses[0]

  for _ in range(1, len(addresses)):

    coordinates = []

    for y in addresses:
      if (y != next_x and not check_exists(y, final_route)):
        coordinate = {
            "pin X": [ next_x[0], next_x[1] ],
            "pin Y": [ y[0], y[1] ] ,
            "datetime": y[2],
            "distance": calculate_distance_haversine(next_x, y),
            "name": y[3]
        }

        if (not check_exists(y, excluded_adresses)):
          coordinates.append(coordinate)

    if (coordinates):
      # Pega a próxima data e horário
      next_datetime = min([ coordinate["datetime"] for coordinate in coordinates ])

      # Cria uma lista de endereços que tem o mesmo horário
      coordinates_by_datetimes = []
      for coordinate in coordinates:
        if ( coordinate["datetime"] == next_datetime):
          coordinates_by_datetimes.append(coordinate)

      # Da lista de endereços que têm o mesmo horário pega o endereço que é mais próximo ao ponto de partida
      minimum_distance_value = min([ coordinate["distance"] for coordinate in coordinates ])
      minimum_distance = None
      for coordinate in coordinates_by_datetimes:
        if ( coordinate["distance"] == minimum_distance_value):
          minimum_distance = coordinate

      # Retorna para o usuário os endereços que não foram escolhidos porque tinham o horário igual
      excluded_adresses += selected_student(minimum_distance, coordinates_by_datetimes)

      final_route.append(minimum_distance)

      # Seta um novo endereço de partida
      next_x = minimum_distance["pin Y"]
    else:
      for address in addresses:
        if (not check_exists(address, final_route) and not check_exists(address, excluded_adresses)):
          next_x = address

  return final_route

def validated_input_date(text):
  datetime_now = datetime.now() + timedelta(hours=1)

  try:
    date = datetime.strptime(text, '%d/%m/%Y %H:%M')
    if (is_valid_date(datetime_now, date)):
      return date
    else:
      print("\nInsira uma data no futuro.")
      return
  except ValueError:
    print("\nTexto inserido está na formatação incorreta.")
    return

def is_valid_date(present_date, future_date):
  if (present_date < future_date):
    return True
  else:
    return False


#===============================================================================
#================================ ENTRADA ======================================
#===============================================================================

# Pega a coordenada geográfica de partida do professor
initial = validated_input(input("Insira a sua localização inicial: "))
datetime_now = datetime.now() + timedelta(hours=1)

# date = datetime.strptime(input('DD/MM/YYYY HH:MM: '), '%d/%m/%Y %H:%M')

if (initial):

  initial.append(datetime_now)
  initial.append("PROFESSOR")

  print("\nDigite o nome do aluno ou digite: \n'Calcular': Para definirmos a sua rota\n'Sair': Para encerrar sem calcular")

  # Pega a coordenada geográfica de cada aluno do dia
  addresses = []
  student = ""
  should_calcule = True
  while student.lower() != 'sair' and student.lower() != 'calcular':
      student = input("\nInsira o nome do aluno: ")
      should_calcule = False if student == 'sair' else True

      if ( student.lower() != "sair" and student.lower() != 'calcular'):

        address = input(f"Insira a localização do {student}: ")
        validated_address = validated_input(address)

        if (validated_address):
          date = validated_input_date(input('Agora insira o a data e horário no formato: DD/MM/YYYY HH:MM '))
          if (date):
            validated_address.append(date)
            validated_address.append(student)
            addresses.append(validated_address)
          else:
            should_calcule = False
            break
        else:
          should_calcule = False
          break

  if (len(addresses) > 0 and should_calcule): 
    # Adiciona a coordenada de partida como primeiro elemento da lista de endereços
    addresses.insert(0, initial)

    #===========================================================================
    #============================ PROCESSAMENTO ================================
    #===========================================================================

    # Ordena a lista de endereços de acordo com a distância
    final_route = ordenate_data(addresses)
    distance_total = sum(route["distance"] for route in final_route)


    #===========================================================================
    #================================= SAÍDA ===================================
    #===========================================================================

    # Formata a saída para o usuário
    print("\nROTA FINAL")
    print_tabulate(final_route)
    print(f"Distância total percorrida: {distance_total:.2f} KM")
  
  print("\nPROGRAMA ENCERRADO.")

# PONTO inicial: -20.285776, -40.2939477
# -20.288459079573457, -40.29302869015847   Júlia   26/05/2021 10:00
# -20.2977177643622, -40.291924050216906    Kássia  26/05/2021 10:00
# -20.31123893978505, -40.28761793085739    Amanda  26/05/2021 11:00
```

```python
import pytz
import math
import tabulate
from datetime import datetime, timedelta
from pycep_correios import get_address_from_cep, WebService
from geopy.geocoders import Nominatim

radius_earth = 6373.0
tz = pytz.timezone('America/Sao_Paulo')

#===============================================================================
#================================ FUNÇÕES ======================================
#===============================================================================

def somalista(number):
    soma = 0
    for i in number:
      soma += i

    return soma

def min_value(values): 
  value_min = values[0]
  for num in values:
    if num < value_min:
      value_min = num
  return value_min 

def get_lat_lon(zip_code):
  '''
    Recebe o CEP como parâmetro
    Retorna: Latitude (float ou None), Longitude (float ou None)
  '''
  only_numbers = zip_code.replace("-", "")
  
  if (len(only_numbers) == 8 and only_numbers.isdigit()):
    address = get_address_from_cep(only_numbers, webservice=WebService.VIACEP)
    geolocator = Nominatim(user_agent="ucl_app")
    location = geolocator.geocode(address['logradouro'] + ", " + address['cidade'] + " - " + address['bairro'])

    if location is not None:
      return location.latitude, location.longitude
    else:
      return None, None
  else:
    return None, None

def print_tabulate(data):
  '''
    Printa a rota final com tabulação
  '''
  # header = data[0].keys()
  header = ["Partida", "Destino", "Aluno", "Horário", "Distância (KM)"]
  rows = []
  for line in data:
    rows.append([ line["pin X"], line["pin Y"], line["name"], line["datetime"], line["distance"] ])
  
  print(tabulate.tabulate(rows, header))
  
def calculate_distance_haversine(pin_1, pin_2):
  '''
    Recebe dois vetores de coordenadas geográficas
    Retorna a distancia entre eles em KM
  '''

  latitude_1 = math.radians(pin_1[0])
  longitude_1 = math.radians(pin_1[1])

  latitude_2 = math.radians(pin_2[0])
  longitude_2 = math.radians(pin_2[1])

  distance_latitude = latitude_2 - latitude_1
  distance_longitude = longitude_2 - longitude_1

  formule_haversine = math.sin(distance_latitude / 2)**2 + math.cos(latitude_1) * math.cos(latitude_2) * math.sin(distance_longitude / 2)**2
  distance_final = 2 * math.atan2(math.sqrt(formule_haversine), math.sqrt(1 - formule_haversine))

  distance_final_km = radius_earth * distance_final

  return distance_final_km

def selected_student(selected_address, addresses):
  '''
    Recebe o endereço escolhido (object) e a lista de endereços (object list)
    Retorna uma lista de endereços que não não iguais ao endereço escolhido
  '''

  excluded_adresses = []
  for address in addresses:
    if (address != selected_address):
      excluded_adresses.append(address)
      print(f"Aluno {address['name']} não adicionado na rota devido a um conflito de horário.")

  return excluded_adresses
  
def check_exists(y, routes):
  '''
    Recebe um vetor de coordenada e verifica se essa coordenada já existe como ponto de partida ou destino
    Retorna True se existe ou False se não existe
  '''

  for route in routes:
    if ((route["pin X"] == [y[0], y[1]]) or (route["pin Y"] == [y[0], y[1]])):
      return True

  return False

def ordenate_data(addresses):
  '''
    Recebe a lista de endereços de entrada pelo usuário
    Retorna a rota final
  '''

  final_route = []
  excluded_adresses = []

  # Endereço de partida
  next_x = addresses[0]
  for _ in range(1, len(addresses)):
    coordinates = []

    # Gera uma lista de coordenadas que ainda não existem na rota final e que também não estejam na lista de endereços excluídos
    for y in addresses:
      if (next_x != y and not check_exists(y, final_route)):
        new_coordinate = {
          "pin X": [ next_x[0], next_x[1] ],
          "pin Y": [ y[0], y[1] ] ,
          "datetime": y[2],
          "distance": calculate_distance_haversine([ next_x[0], next_x[1] ], [ y[0], y[1] ]),
          "name": y[3]
        }

        if (not check_exists(y, excluded_adresses)):
          coordinates.append(new_coordinate)

    # Se a lista não está vazia faz os calculos
    if (coordinates):
      # Pega o menor horário da lista de endereços
      next_datetime = min_value([ coordinate["datetime"] for coordinate in coordinates])

      # Cria uma lista de endereços que tem o mesmo horário
      # coordinates_by_datetimes = [ coordinate for coordinate in coordinates if coordinate["datetime"] == next_datetime]
      for coordinate in coordinates[:]:
        if ( coordinate["datetime"] != next_datetime ):
          coordinates.remove(coordinate)

      # Da lista de endereços que têm o mesmo horário pega o endereço que é mais próximo ao ponto de partida
      minimum_distance_value = min_value([ coordinate["distance"] for coordinate in coordinates ])
      minimum_distance = None
      for coordinate in coordinates:
        if ( coordinate["distance"] == minimum_distance_value):
          minimum_distance = coordinate

      # Retorna para o usuário os endereços que não foram escolhidos porque tinham o horário igual
      excluded_adresses += selected_student(minimum_distance, coordinates)

      # Adiciona o endereço na rota final
      final_route.append(minimum_distance)

      # Seta um novo endereço de partida
      next_x = minimum_distance["pin Y"]

    # Se a lista está vazia verifica se há um possível novo ponto de partida
    else:
      for address in addresses:
        if (not check_exists(address, final_route) and not check_exists(address, excluded_adresses)):
          next_x = address

  return final_route

def validated_input_date(text):
  datetime_now = datetime.now(tz) + timedelta(hours=1)

  # try:
  #   date = datetime.strptime(text, '%d/%m/%Y %H:%M')
  #   if (is_valid_date(datetime_now.replace(tzinfo=None), date)):
  #     return date
  #   else:
  #     print("\nInsira uma data no futuro.")
  #     return
  # except ValueError:
  #   print("\nTexto inserido está na formatação incorreta.")
  #   return

  datetime_text = text.split(" ")
  if (len(datetime_text) != 2):
    print("Insira a data e o horário")
    return
  else:
    date_numbers = datetime_text[0].split("/")
    time_numbers = datetime_text[1].split(":")

    if (len(date_numbers) != 3 or not (all_int(date_numbers)) ):
      print("Data em formato inválido")
      return
    elif (len(time_numbers) != 2 or not (all_int(time_numbers)) ):
      print("Horário em formato inválido")
      return
    else:
      day, month, year = date_numbers
      hour, minute = time_numbers
      if not (((int(day) >= 1 and int(day) <= 31) and (int(month) >= 1 and int(month) <= 12) and int(year) >= 1000)):
        print("Data inválida")
        return
      elif (not ( int(hour) >= 0 and ( int(minute) >= 0 and int(minute) <= 59 ))):
        print("Horário inválido")
        return
      else:
        date = datetime(int(year), int(month), int(day), int(hour), int(minute), tzinfo=tz)
        if (is_valid_date(datetime_now, date)):
          return date
        else:
          print("\nInsira uma data no futuro.")
          return

def is_valid_date(present_date, future_date):
  if (present_date < future_date):
    return True
  else:
    return False

def all_int(vector):
  is_all_int = True
  for item in vector:
    if ( not item.isdigit() ):
      is_all_int = False

  return is_all_int

#===============================================================================
#================================ ENTRADA ======================================
#===============================================================================

# Pega a coordenada geográfica de partida do professor
zip_code = input("Insira o CEP do seu endereço de partida >> ")
datetime_now = datetime.now(tz) + timedelta(hours=1)

# date = datetime.strptime(input('DD/MM/YYYY HH:MM: '), '%d/%m/%Y %H:%M')
latitude, longitude = get_lat_lon(zip_code)

if (latitude is not None and longitude is not None):
  initial = [ latitude, longitude, datetime_now, "Professor" ]

  print("\nDigite o nome do aluno ou digite: \n'Calcular': Para definirmos a sua rota\n'Sair': Para encerrar sem calcular")

  # Pega a coordenada geográfica de cada aluno do dia
  addresses = []
  student = ""
  should_calcule = True
  while student.lower() != 'sair' and student.lower() != 'calcular':
      student = input("\nInsira o nome do aluno >> ")
      should_calcule = False if student == 'sair' else True

      if ( student.lower() != "sair" and student.lower() != 'calcular'):
        latitude = None
        longitude = None
        
        address = input(f"Insira o CEP de {student}: ")
        latitude, longitude = get_lat_lon(address)

        if (latitude is not None and longitude is not None):
          date = validated_input_date(input('Agora insira o a data e horário no formato: DD/MM/YYYY HH:MM >> '))
          if (date):
            addresses.append([ latitude, longitude, date, student ]) 
          else:
            should_calcule = False
            break
        else:
          should_calcule = False
          break

  if (len(addresses) > 0 and should_calcule): 
    # Adiciona a coordenada de partida como primeiro elemento da lista de endereços
    addresses.insert(0, initial)

    #===========================================================================
    #============================ PROCESSAMENTO ================================
    #===========================================================================

    # Ordena a lista de endereços de acordo com a distância
    final_route = ordenate_data(addresses)

    # Calcula a distância percorrida
    lista_distancias = [route["distance"] for route in final_route]
    distance_total = somalista(lista_distancias)


    #===========================================================================
    #================================= SAÍDA ===================================
    #===========================================================================

    # Formata a saída para o usuário
    print("\nROTA FINAL")
    print_tabulate(final_route)
    print(f"Distância total percorrida: {distance_total:.2f} KM")
  
  print("\nPROGRAMA ENCERRADO.")

  # 29060020
  # Aluno 1         29055245     09/06/2021 12:00
  # Aluno 2         29060120     09/06/2021 15:00
  # Aluno 3         29060770     09/06/2021 17:00
```
