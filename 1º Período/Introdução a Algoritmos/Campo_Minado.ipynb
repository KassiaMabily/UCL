## Versão 1
Simplificada, só abre a vizinhança em volta da célula escolhida

**PRONTA PARA RODAR**


```python
import random
from tabulate import tabulate

MINE = '*';
CLOSED = '-';
EMPTY = ' ';

SIZE = 5
REAL_BOARD  = []
MY_BOARD   = []


def initialize(): 
  global REAL_BOARD, MY_BOARD

  REAL_BOARD  = [ [0] * SIZE for i in range(SIZE)]
  MY_BOARD    = [ [CLOSED] * SIZE for i in range(SIZE)] 


def print_board(board):
  headers = [ i for i in range(SIZE) ]
  print(tabulate(board, headers = headers, showindex = headers))


def place_mines(mines):

  for _ in range(mines + 1):
    row = random.randint(1, SIZE) - 1
    col = random.randint(1, SIZE) - 1
    REAL_BOARD[row][col] = MINE


def count_open():
  count = 0
  for i in range(SIZE):
    for j in range(SIZE):
      if MY_BOARD[i][j] != CLOSED:
        count += 1
  return count


def generate_neighbors():
  for r in range(0, SIZE):
    for c in range(0, SIZE):
      for rr in range(r - 1, r + 2):
        for cc in range(c - 1, c + 2):
          if ((rr >= 0 and rr < SIZE) and (cc >= 0 and cc < SIZE)):
            if (  REAL_BOARD[rr][cc] == MINE  ):
              if ((r >= 0 and r < SIZE) and (c >= 0 and c < SIZE)):
                if (  REAL_BOARD[r][c] != MINE  ):
                  REAL_BOARD[r][c] += 1

def open_empty(line, column):
  for r in range(line - 1, line + 2):
    for c in range(column - 1, column + 2):
      if ((r >= 0 and r < SIZE) and (c >= 0 and c < SIZE)):
        if ( REAL_BOARD[r][c] != MINE ):
          MY_BOARD[r][c] = REAL_BOARD[r][c]

def play_game():
  moves_left = ( SIZE * SIZE ) - MINES
  count_is_open = count_open()

  print_board(MY_BOARD)
  while (count_is_open < moves_left):
    l = int(input("Linha: "))
    c = int(input("Coluna: "))
            
    if (MY_BOARD[l][c] != CLOSED):
      print("\nEsta posicao ja esta aberta!\n");
    elif (REAL_BOARD[l][c] == MINE):
      MY_BOARD[l][c] = MINE
      print("\nVocê atingiu uma bomba!\n");
      print_board(REAL_BOARD)
      break;
    elif (REAL_BOARD[l][c] != MINE):
      # Abre a célula
      MY_BOARD[l][c] = REAL_BOARD[l][c]

      # Para abrir uma casa por vez, comentar a função abaixo
      open_empty(l,c)

      # Conta a quantidade de celulas abertas
      count_is_open = count_open()
      
      print()
      print_board(MY_BOARD)
      print()
    else:
      MY_BOARD[l][c] = REAL_BOARD[l][c]
      print()
      print_board(MY_BOARD)
      print()
    
  if (moves_left == count_is_open):
    print("\nParabens! Voce ganhou!\n")


def main():
    mines = 5

    initialize()
    
    place_mines(mines)

    generate_neighbors()

    # Para mostrar a resposta
    print_board(REAL_BOARD)

    play_game()

main()
```

Recursão para abrir a vizinha em volta da célula escolhida até as minas

```python
import random
from tabulate import tabulate
 
SIZE  = 0
MINES = 0

REAL_BOARD  = []
MY_BOARD    = [] 

def choose_level():
  print("Aperte 1 para INICIANTE (9 * 9)")
  print("Aperte 2 para ITERMEDIÁRIO (16 * 16)")
  print("Aperte 3 para AVANÇADO (24 * 24)")

  global SIZE, MINES

  level = int(input("Escolha a dificuldade >> "))

  if level == 1:
    SIZE = 9
    MINES = 10

  if level == 2:
    SIZE = 16
    MINES = 40

  if level == 3:
    SIZE = 24
    MINES = 99


def is_valid(row, col):
  return (row >= 0) and (row < SIZE) and (col >= 0) and (col < SIZE)


def is_mine (row, col, board):
    if (board[row][col] == '*'):
        return True;
    else:
        return False

def is_all_int(vector):
  valid = True
  for item in vector:
    if ( not item.isdigit() ):
      valid = False

  return valid


def initialize(): 
  global REAL_BOARD, MY_BOARD

  REAL_BOARD  = [ [0] * SIZE for i in range(SIZE)]
  MY_BOARD    = [ [0] * SIZE for i in range(SIZE)] 

  for i in range(SIZE):
    for j in range(SIZE):
      MY_BOARD[i][j] = '-'
      REAL_BOARD[i][j] = '-'


def place_mines():
  for _ in range(MINES + 1):
    row = random.randint(1, SIZE) - 1
    col = random.randint(1, SIZE) - 1
    REAL_BOARD[row][col] = '*'


def print_board(board):
  headers = [ i for i in range(SIZE) ]
  print(tabulate(board, headers = headers, showindex = headers))


def make_move():
  pos = []
  valid = False

  while len(pos) != 2 or not valid:
    pos = input("Digite: linha coluna >> ").split(" ")

    if len(pos) == 2 and is_all_int(pos):
      if ( is_valid(int (pos[0]), int (pos[1])) ):
        valid = True
      elif ( int (pos[0]) == -1 and int (pos[1]) == -1):
        valid = True
      else:
        print(f"Entrada inválida. Digite apenas números entre 0 e {SIZE - 1}")
    else:
      print("Entrada inválida. Digite apenas 2 números")

  return int (pos[0]), int (pos[1])


def count_mines_neighbors(row, col):
  count = 0

  if (is_valid (row - 1, col) == True):
    if (is_mine (row - 1, col, REAL_BOARD) == True):
      count +=1
  
  if (is_valid (row + 1, col) == True):
    if (is_mine (row + 1, col, REAL_BOARD) == True):
      count+=1

  if (is_valid (row, col + 1) == True):
    if (is_mine (row, col + 1, REAL_BOARD) == True):
      count+=1
  
  if (is_valid (row, col - 1) == True):
    if (is_mine (row, col - 1, REAL_BOARD) == True):
      count+=1

  if (is_valid (row - 1, col + 1) == True):
    if (is_mine (row - 1, col + 1, REAL_BOARD) == True):
      count+=1
  
  if (is_valid (row - 1, col - 1) == True):
    if (is_mine (row - 1, col - 1, REAL_BOARD) == True):
      count+=1

  if (is_valid (row + 1, col + 1) == True):
    if (is_mine (row + 1, col + 1, REAL_BOARD) == True):
      count+=1

  if (is_valid (row + 1, col - 1) == True):
    if (is_mine (row + 1, col - 1, REAL_BOARD) == True):
      count+=1

  return count


def play_cell(row, col, moves_left):
  if (MY_BOARD[row][col] != '-'):
    return False, moves_left

  if (REAL_BOARD[row][col] == '*'):
    MY_BOARD[row][col] = '*'

    for i in range(SIZE):
      for j in range(SIZE):
        if ( REAL_BOARD[i][j] == "*" ):
          MY_BOARD[i][j] = '*'

    print("\nVOCÊ PERDEU!\n")
    print_board(MY_BOARD)
    
    return True, moves_left

  else:
    count_mines = count_mines_neighbors(row, col)
    moves_left -= 1

    MY_BOARD[row][col] = count_mines if count_mines > 0 else 0


    '''
    1 | 2 | 3
    4 | p | 5
    6 | 7 | 8
    '''
    if not count_mines:   
      # 1   
      if (is_valid (row - 1, col - 1) == True):
        if (is_mine (row - 1, col - 1, REAL_BOARD) == False):
          play_cell(row-1, col-1, moves_left)

      # 2
      if (is_valid (row - 1, col) == True):
        if (is_mine (row - 1, col, REAL_BOARD) == False):
          play_cell(row-1, col, moves_left)

      # 3
      if (is_valid (row-1, col+1) == True):
        if (is_mine (row-1, col+1, REAL_BOARD) == False):
          play_cell(row-1, col+1, moves_left)

      #4
      if (is_valid (row, col - 1) == True):
        if (is_mine (row, col - 1, REAL_BOARD) == False):
          play_cell(row, col+1, moves_left)

      # 5
      if (is_valid (row, col+1) == True):
        if (is_mine (row, col+1, REAL_BOARD) == False):
          play_cell(row, col+1, moves_left)

      # 6
      if (is_valid (row+1, col-1) == True):
        if (is_mine (row+1, col-1, REAL_BOARD) == False):
          play_cell(row+1, col-1, moves_left)

      # 7
      if (is_valid (row+1, col) == True):
        if (is_mine (row+1, col, REAL_BOARD) == False):
          play_cell(row+1, col, moves_left)

      # 8
      if (is_valid (row+1, col+1) == True):
        if (is_mine (row+1, col+1, REAL_BOARD) == False):
          play_cell(row+1, col+1, moves_left)
 
    return False, moves_left

  
def play():
  
  game_over = False

  x = y = 0

  initialize()

  place_mines()

  # print_board(REAL_BOARD)

  moves_left = SIZE * SIZE - MINES

  print("Para sair digite: -1 -1")

  while ( game_over == False and ( x != -1 and y != -1 ) ):
    print("\nTABULEIRO ATUAL: ")
    print_board(MY_BOARD)

    x, y = make_move()

    if ( x != -1 and y != -1 ):
      game_over, moves_left = play_cell(x, y, moves_left)

      if ((game_over == False) and (moves_left == 0)):
        print("VOCÊ GANHOU!!!")
        game_over = True
    else:
      print("JOGO ENCERRADO")

def main():
  choose_level()
  play()

main()
```
